```python
import pandas as pd
import numpy as np
```

```python
sim_outputs = pd.read_csv("/Users/sara.hippmann/Downloads/PRICE-519 Field Goal Market Investigation - Sheet1.csv")
sim_outputs.drop(columns = ["Touchodwns Home", "Fields Home", "Touchdowns Away", "Fieldgoals Away"], inplace=True)
```

```python
sim_outputs["HomeSplit"] = 0
sim_outputs["AwaySplit"] = 0
sim_outputs
```

```python
def solve_equation(N):
    solutions = []
    for y in range(N // 7 + 1):
        if (N - 7*y) % 3 == 0:   # check divisibility
            x = (N - 7*y) // 3
            solutions.append((x, y))
    return solutions
```

```python
sim_outputs["HomeSplit"] = sim_outputs["HomePoints"].apply(solve_equation)
sim_outputs["AwaySplit"] = sim_outputs["AwayPoints"].apply(solve_equation)
```

```python
sim_outputs.to_csv("FG Split.csv")
```
